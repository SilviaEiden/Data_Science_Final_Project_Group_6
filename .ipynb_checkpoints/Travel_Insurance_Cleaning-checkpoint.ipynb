# Import dependencies

```python
# Dependencies
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
from config import db_password
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

# Dataset cleaning

```python
# Import our input dataset
travel_ins_df = pd.read_csv('Resources/travel insurance.csv')
print(travel_ins_df.shape)
travel_ins_df.head()
```

```python
# Checking datatypes
travel_ins_df.dtypes
```

```python
# Checking for nulls
travel_ins_df.isnull().sum()
```

```python
# Drop Gender column
travel_ins_df.drop(columns=["Gender"], inplace=True)
```

```python
# Claim value counts
claim = travel_ins_df["Claim"].value_counts()
claim
```

```python
# Transform string column
def change_string(claim):
    if claim == "Yes":
        return 1
    else:
        return 0

travel_ins_df["Claim"] = travel_ins_df["Claim"].apply(change_string)
```

```python
# Check Duration column if any negative
duration_neg = travel_ins_df["Duration"].sort_values(ascending=True)
duration_neg.head(70)
```

```python
# Remove negative values of the Duration column
travel_ins_df = travel_ins_df.loc[travel_ins_df["Duration"] > 0]
```

```python
# Check Duration column on max data
duration = travel_ins_df["Duration"].sort_values(ascending=False)
duration.head(20)
```

```python
# Create a boxplot for Duration
fig, ax = plt.subplots(figsize=(4,10))
travel_ins_df.boxplot(column='Duration', return_type='axes')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
```

```python
# Remove Duration > 600 outliers
travel_ins_df = travel_ins_df.loc[travel_ins_df["Duration"] < 600]
```

```python
# Create a boxplot for Duration to check
fig, ax = plt.subplots(figsize=(4,10))
travel_ins_df.boxplot(column='Duration', return_type='axes')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
```

```python
# Check Age column if any negative
age_neg = travel_ins_df["Age"].sort_values(ascending=True)
age_neg.head()
```

```python
# Check Age column on max data
age = travel_ins_df["Age"].sort_values(ascending=False)
age.head()
```

```python
# Get the number of ages
age_count = travel_ins_df.groupby(["Age"]).count()["Claim"]
age_count
```

```python
# Create a boxplot for Age
fig, ax = plt.subplots(figsize=(4,10))
travel_ins_df.boxplot(column='Age', return_type='axes')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
```

```python
# Remove 118 years outlier
travel_ins_df = travel_ins_df.loc[travel_ins_df["Age"] < 118]
```

```python
# Create a boxplot for Age to check
fig, ax = plt.subplots(figsize=(4,10))
travel_ins_df.boxplot(column='Age', return_type='axes')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
```

```python
# Create a new DataFrame that holds only Net Sales and Commision
money_df = travel_ins_df.copy()
money_df = money_df.filter(items=["Net Sales", "Commision (in value)"])

print(money_df.shape)
money_df.head()
```

```python
# Drop the Net Sales and Commision columns
travel_ins_df.drop(columns=["Net Sales", "Commision (in value)"], inplace=True)

print(travel_ins_df.shape)
travel_ins_df.head()
```

# Export dataset

```python
# Create the connection to the PostgreSQL database
try:
    db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/travel_insurance"
    engine = create_engine(db_string)
except:
    db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5433/travel_insurance"
    engine = create_engine(db_string)

# Function to import to SQL
def to_sql(dataframe, name):
    dataframe.to_sql(name=name, con=engine, method='multi', if_exists='replace')
```

```python
# Saving data
travel_ins_df.to_csv('Resources/travel_insurance.csv', index=False)

to_sql(travel_ins_df, "travel")
to_sql(money_df, "money")
```

# Correlation and Heatmap

## Correlation and Heatmap with LabelEncoder

```python
# Our input dataset
le_corr_heatmap_df = travel_ins_df.copy()
print(le_corr_heatmap_df.shape)
le_corr_heatmap_df.head()
```

```python
# Apply encoder
columns_list = le_corr_heatmap_df.dtypes.loc[lambda x: x == "object"].index.tolist()
for column in columns_list:
    le_corr_heatmap_df[column] = LabelEncoder().fit_transform(le_corr_heatmap_df[column])
```

```python
print(le_corr_heatmap_df.shape)
le_corr_heatmap_df.head()
```

```python
# Correlation dataframe
le_corr_heatmap_df.corr()
```

```python
# Heatmap #1
plt.figure(figsize=(8, 6))
heatmap_1 = sns.heatmap(le_corr_heatmap_df.corr(), vmin=-1, vmax=1, annot=True)
heatmap_1.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=16)

plt.savefig("Analysis/heatmap_1.png", dpi=400, bbox_inches = "tight")
plt.show()
```

```python
# Heatmap #2
plt.figure(figsize=(4, 6))
heatmap_2 = sns.heatmap(le_corr_heatmap_df.corr()[['Claim']].sort_values(by='Claim', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap_2.set_title('Features Correlating with Claim', fontdict={'fontsize':16}, pad=14)

plt.savefig("Analysis/heatmap_2.png", dpi=400, bbox_inches = "tight")
plt.show()
```

## Correlation and Heatmap with 'get_dummies' encoder

```python
# Our input dataset
dumm_corr_heatmap_df = travel_ins_df.copy()
print(dumm_corr_heatmap_df.shape)
dumm_corr_heatmap_df.head()
```

```python
# Agency value counts
agency = dumm_corr_heatmap_df["Agency"].value_counts()
agency
```

```python
# Determine which values to replace
replace_agency = list(agency[agency < 500].index)

# Replace in dataframe
for agencies in replace_agency:
    dumm_corr_heatmap_df["Agency"] = dumm_corr_heatmap_df["Agency"].replace(agencies,"Other Agencies")
    
# Check to make sure binning was successful
dumm_corr_heatmap_df["Agency"].value_counts()
```

```python
# Product Name value counts
prod_name = dumm_corr_heatmap_df["Product Name"].value_counts()
prod_name
```

```python
# Determine which values to replace
replace_prod_name = list(prod_name[prod_name < 1000].index)

# Replace in dataframe
for products in replace_prod_name:
    dumm_corr_heatmap_df["Product Name"] = dumm_corr_heatmap_df["Product Name"].replace(products,"Other Plans")
    
# Check to make sure binning was successful
dumm_corr_heatmap_df["Product Name"].value_counts()
```

```python
# Destination value counts
destination = dumm_corr_heatmap_df["Destination"].value_counts()
destination
```

```python
# Determine which values to replace
replace_destination = list(destination[destination < 1000].index)

# Replace in dataframe
for destinations in replace_destination:
    dumm_corr_heatmap_df["Destination"] = dumm_corr_heatmap_df["Destination"].replace(destinations,"Other Destinations")
    
# Check to make sure binning was successful
dumm_corr_heatmap_df["Destination"].value_counts()
```

```python
# Encode Labels with 'get_dummies'
dumm_corr_heatmap_df = pd.get_dummies(dumm_corr_heatmap_df,
                                      columns=dumm_corr_heatmap_df.dtypes.loc[lambda x: x == "object"].index.tolist())
print(dumm_corr_heatmap_df.shape)
dumm_corr_heatmap_df.head()
```

```python
# Correlation dataframe
dumm_corr_heatmap_df.corr()
```

```python
# Heatmap #3
plt.figure(figsize=(12, 10))
heatmap_3 = sns.heatmap(dumm_corr_heatmap_df.corr(), vmin=-1, vmax=1)
heatmap_3.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=16)

plt.savefig("Analysis/heatmap_3.png", dpi=400, bbox_inches = "tight")
plt.show()
```

```python
# Heatmap #4
plt.figure(figsize=(8, 12))
heatmap_4 = sns.heatmap(dumm_corr_heatmap_df.corr()[['Claim']].sort_values(by='Claim',
                                                        ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap_4.set_title('Features Correlating with Claim', fontdict={'fontsize':18}, pad=16)

plt.savefig("Analysis/heatmap_4.png", dpi=400, bbox_inches = "tight")
plt.show()
```
